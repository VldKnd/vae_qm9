```python
def show_mol(mol):
    plt.figure(figsize=(5, 6))
    
    img = Chem.Draw.MolToImage(Chem.MolFromSmiles(mol))

    title = mol
    plt.title(title)
    plt.axis("off")
    plt.imshow(img)

    plt.show()
    
def show_mols(mols, rows = None, cols = None):
    ## Assertion of correct input smiles strings
    for s in mols:
        assert Chem.MolFromSmiles(s) is not None, "The smiles strings have to be valid"
        
    if rows is None and cols is None:
        cols = 3
        rows = math.ceil(len(mols)/cols)
        
    elif rows is None:
        rows = math.ceil(len(mols)/cols)
        
    elif cols is None:
        cols = math.ceil(len(mols)/rows)
    
    ## Assertion of correct figures grid (check matplotlib.pyplot subplot gird)
    assert rows*cols >= len(mols), "The grid for images has to fit all the figures"
    
    
    ## Main part
    plt.figure(figsize=(20, 6*rows))
    for i, mol in enumerate(mols):
        position = 1 + i
        img = Chem.Draw.MolToImage(Chem.MolFromSmiles(mol))

        ax = plt.subplot(rows, cols, position)
        
        title = mol
        ax.set_title(title)
        ax.axis("off")
        ax.imshow(img)

    plt.show()
```
